```python
#Problem Statement 1: Employee Performance Bonus Eligibility

employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
highest_score=max(employees.values())
top_performer=[]
for name,score in employees.items():
    if score==highest_score:
        top_performer.append(name)
print(f"Top Performers Eligible for Bonus:{top_performer}score:({highest_score})")
```

```python
#Problem Statement 2.Search Query Keyword Analysis

import string
sentence="Buy mobile phone buy phone online"
lowercase=sentence.lower()
query=lowercase.translate(str.maketrans('','',string.punctuation))
words=query.split()
counts={}
for word in words:
    if word in counts:
        counts[word]+=1
    else:
        counts[word]=1
result={}
for word,count in counts.items():
    if count>1:
        result[word]=count
print(result)
```

```python
# Problem Statement 3.Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_reading=[]
for hour_index, reading_value in enumerate(sensor_readings):
    if reading_value %2==0:
        valid_reading.append( (hour_index, reading_value))
print("Valid Sensor Readings (Hour, Value):")
print(valid_reading)
```

```python
#Problem Statement 5.Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]
average=sum(sales)/len(sales)
limits=average*1.3
for i in range(len(sales)):
    if sales[i]>limits:
        print("Day",i+1,":",sales[i])
```

```python
#Problem Statement 6.Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
count={}
for user in user_ids:
    if user in count:
        count[user]+=1
    else:
        count[user]=1
for user,times in count.items():
    if times>1:
        print(user,"->",times,"times")
```

1. User Login Check

```python
username="admin"
password="1234"

input_username=input("Enter username: ")
input_password=input("Enter password: ")

if input_username==username and input_password==password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2. Pass / Fail Analyzer

```python
marks = list(map(int,input().split(",")))

pass_count=0
fail_count=0

for i in marks:
  if i>=50:
    pass_count+=1
  else:
    fail_count+=1

print("-------Final Result-------")
print("Total Students:",len(marks))
print("Pass students: ",pass_count)
print("Fail students: ",fail_count)
```

3. Simple Data Cleaner

```python
names=list(map(str,input().split(",")))
cleaned_names=[]
for name in names:
  cleaned=name.strip()                                # remove outer spaces
  cleaned=cleaned.replace('"','').replace("'","")     # remove double and single quotes
  cleaned=cleaned.strip()                             # remove spaces again
  cleaned=cleaned.lower()                             # convert to lowercase
  cleaned_names.append(cleaned)

print("["+",".join(f'"{name}"' for name in cleaned_names)+"]")
```

4. Message Length Analyzer

```python
msgs=list(map(str,input().split(",")))

for msg in msgs:
  length=len(msg)
  print("Message:",msg)
  print("Length:",length)

  if length>10:
    print("This message is longer than 10 characters")

  print()
```

5. Error Message Detector

```python
logs=list(map(str,input().split(",")))

count=0

for log in logs:
  cleaned=log.strip()               # remove spaces
  cleaned=cleaned.replace('"','')   # remove quotes
  cleaned=cleaned.replace("'","")   # remove single quotes

  if cleaned=="ERROR":
    count+=1

print("Total Errors:",count)
```
